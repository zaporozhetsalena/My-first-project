–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é calculation —è–∫–∞ –ø—Ä–∏–π–º–∞—î –Ω–∞ –≤—Ö—ñ–¥ 4 –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —Ç–∏–ø—É int - a, b, c, d —ñ –≤–µ—Ä—Ç–∞—î —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Ç–∞–∫–æ—ó —Ñ–æ—Ä–º—É–ª–∏:  (a+b‚àíc)‚àód .
–í–∏–∫–ª–∏—á—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–æ–º—É –Ω–∞–±–æ—Ä—ñ –∞—Ä–≥—É–º–µ–Ω—Ç—ñ–≤ —ñ –≤–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω:

a = 16
b = 4
c = 9
d = 3

```python
def calculation(a:int, b:int, c:int, d:int) -> int:
    assert type(a) == int and type(b) == int and type(c) == int and type(d) == int
    return (a+b-c)*d
```

```python
calculation(16,4,9,3)
```

–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é multipy_numbers, —è–∫–∞ –ø—Ä–∏–π–º–∞—î –¥–≤–∞ —á–∏—Å–ª–∞ —è–∫ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —ñ –ø–æ–≤–µ—Ä—Ç–∞—î —ó—Ö –¥–æ–±—É—Ç–æ–∫.
–í–∏–∫–ª–∏—á—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é multipy_numbers –Ω–∞ –ø–∞—Ä–∞—Ö —á–∏—Å–µ–ª

10 —ñ 2

7 —ñ 8

1024 —ñ 999.

–í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫–æ–∂–Ω–æ–≥–æ –≤–∏–∫–ª–∏–∫—É. –ü—Ä–∏ –≤–∏–∫–æ–Ω–∞–Ω–Ω—ñ —Ü—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º —Ü–∏–∫–ª—É - –æ—Ç—Ä–∏–º–∞—î—Ç–µ 2—Ö –±–∞–ª—ñ–≤ –∑–∞ –Ω—å–æ–≥–æ :)

```python
def multipy_numbers(a, b):
    return a * b

pairs = [(10, 2), (7, 8), (1024, 999)]

new_list = [multipy_numbers(a, b) for a, b in pairs]

print(new_list)
```

–î—ñ–ª–µ–Ω–Ω—è –∑ –ø–µ—Ä–µ–≤—ñ—Ä–∫–æ—é —Ç–∏–ø—ñ–≤.

–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é int_division —è–∫–∞ –ø—Ä–∏–π–º–∞—î –Ω–∞ –≤—Ö—ñ–¥ –¥–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —Ç–∏–ø—É int - a —ñ b, —ñ –≤–∏–∫–æ–Ω—É—î –¥—ñ–ª–µ–Ω–Ω—è –ø–µ—Ä—à–æ–≥–æ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ –Ω–∞ –¥—Ä—É–≥–∏–π: a/b. –ü—Ä–∏ —Ü—å–æ–º—É —Ñ—É–Ω–∫—Ü—ñ—è –ø–µ—Ä–µ–¥ –¥—ñ–ª–µ–Ω–Ω—è–º –ø–µ—Ä–µ–≤—ñ—Ä—è—î, —á–∏ –¥—Ä—É–≥–∏–π –∞—Ä–≥—É–º–µ–Ω—Ç –Ω–µ –¥–æ—Ä—ñ–≤–Ω—é—î 0. –Ø–∫—â–æ –¥—Ä—É–≥–∏–π –∞—Ä–≥—É–º–µ–Ω—Ç –Ω–µ 0, —Ñ—É–∫–Ω—Ü—ñ—è –≤–µ—Ä—Ç–∞—î —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥—ñ–ª–µ–Ω–Ω—è, —ñ–Ω–∞–∫—à–µ - –∑–∞–º—ñ—Å—Ç—å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –¥—ñ–ª–µ–Ω–Ω—è, –≤–µ—Ä—Ç–∞—î –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –Ω–∞ –µ–∫—Ä–∞–Ω "–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0".

–Ø–∫—â–æ —Ç–∏–ø–∏, –Ω–∞–¥–∞–Ω—ñ —É —Ñ—É–Ω–∫—Ü—ñ—é, –Ω–µ `int` - —Ñ—É–Ω–∫—Ü—ñ—è –≤–∏–≤–æ–¥–∏—Ç—å –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.".
–ó–∞–ø—É—Å—Ç—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö –ø–∞—Ä–∞—Ö —á–∏—Å–µ–ª:
12 —ñ 6

20 —ñ 0

112 —ñ 10

99 —ñ 0

11 i 1.1

—ñ –≤–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó: " - —ñ –¥–∞–ª—ñ —Ç–µ, —â–æ –ø–æ–≤–µ—Ä–Ω—É–ª–∞ —Ñ—É–Ω–∫—Ü—ñ—è.

–î–∞–π—Ç–µ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ –ø–∏—Ç–∞–Ω–Ω—è: "–ß–æ–º—É –≤ –¥–µ—è–∫–∏—Ö –≤–∏–ø–∞–¥–∫–∞—Ö —Ñ—É–Ω–∫—Ü—ñ—è –ø–æ–≤–µ—Ä–Ω—É–ª–∞ None?"

```python
def int_division(a, b):
    if not isinstance(a, int) or not isinstance(b, int):
        print("–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.")
        return
    if b == 0:
        print("–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0")
        return
    return a / b


pairs = [(12, 6), (20, 0), (112, 10), (99, 0), (11, 1.1)]

for a, b in pairs:
    result = int_division(a, b)
    print("–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó:", result)
    
```
