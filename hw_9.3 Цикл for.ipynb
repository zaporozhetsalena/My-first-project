# # # Задайте список numbers, який містить значення від 0 до 9 включно. Використовуючи цикл for, обчисліть суму елементів в створеному списку та виведіть фінальну суму на екран.

```python
numbers = list(range(10))
total = 0

for i in numbers:
    total += i

print(total)
```

# # # Використовуючи цикл  for виведіть для кожного елементу списку numbers,який ми створили в завданні 1, чи цей елемент дорівнює 5. Тобто після виконання циклу на екрані має бути 10 bool значень.

```python
numbers = list(range(10))
for i in numbers:
    print(i == 5)
```

# # # Використовуючи цикл  for для кожного елементу списку numbers, який ми створили в завданні 1, виявіть, чи цей елемент ділиться на 2 без остачі, та запишіть результат у новий список numbers_divisible_by_2.
Увага! Виконання завдання відрізняється від попереднього тим, що тепер ми результат обчислення на кожній ітерації записуємо у новий список. До того ж ми тут виконуємо іншу математичну операцію.

Виведіть фільнальний список numbers_divisible_by_2 на екран. Очікуваний результат: [True, False, True, False, True, False, True, False, True, False]

```python
numbers = list(range(10))
numbers_divisible_by_2 = []

for i in numbers:
    numbers_divisible_by_2.append(i % 2 == 0)

print(numbers_divisible_by_2)
```

```python
numbers_divisible_by_2 = [n % 2 == 0 for n in numbers]
print(numbers_divisible_by_2)
```

# # # Задано два списки в клітинці нижче - a і b. Створіть цикл, результатом якого буде список result, що містить елементи типу bool і який фактично каже, "чи на цьому індексі елементи наших двох списків співпадають". Приклад:
a = [1, 2, 3, 6]
b = [1, 7, 3, 6]
# ми виконали поелементне порівняння двох списків і отримали новий список result
# result містить наступний результат -> [True, False, True, True] В кінці виведіть вміст result на екран.

```python
a = [1, 2, 3, 6]
b = [1, 7, 3, 6]

result = []

for i in range(len(a)):
    result.append(a[i] == b[i])

print(result)
```

Знайдіть суму всіх елементів списку numbers та запишіть її у змінну numbers_sum. Можна з використанням циклу, а можна і без.

Обчисліть numbers_mean шляхом ділення суми елементів у списку на їх кількість. Кількість елементів в списку - це його довжина.

Виведіть numbers_mean, округлений до 2 цифр після коми, на екран.

Підказка: округлити число можна з використанням вбудованої функції Python round(). numbers = [42, 51, 29, 2, 13, 44, 80, 66, 16] Очікуваний результат: 38.11

```python
numbers = [42, 51, 29, 2, 13, 44, 80, 66, 16] 
numbers_sum = sum(numbers)
numbers_count = len(numbers)
numbers_mean = numbers_sum / numbers_count
numbers_mean = round(numbers_mean, 2)
print(numbers_mean)
```
