3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = pd.pivot_table(df, index="Branch", columns="Product line",values=["Invoice ID"],aggfunc="count",fill_value=0)
```

Досить цікаве розподілення продуктової лінійки серед міст, тому виходячи з цього нова гіпотеза 2 - чи впливає гендер на покупку тих чи інших товарів?

```python
gender_category = df.groupby(['Gender', 'Product line'])['Invoice ID'] \
                    .count() \
                    .unstack(fill_value=0)
```

```python
gender_category_percent = gender_category.div(gender_category.sum(axis=1), axis=0)
```

```python
print(gender_category_percent.round(2))
```

Цікава статистика: Health and beauty - 0,18 більше купують чоловіки. І в Fashion accessories більше купують жінки. Також Sports and travel більше покупок зі сторони жінок.

Ще можна дослідити чивідрізняються чоловіки та жінки за способом оплати

```python
gender_payment_percent = df.groupby('Gender')['Payment'].value_counts(normalize=True).unstack(fill_value=0)
```

```python
print(gender_payment_percent)
```
